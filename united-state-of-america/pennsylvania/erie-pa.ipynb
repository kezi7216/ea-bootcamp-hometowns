## The U.S. Brig Niagara

This landmark in Erie, PA is a replica of the battleship that participated in the Battle of Lake Erie during the War of 1812.  It was the relief flagship for Oliver Hazard Perry after his main flagship was destroyed during the battle.  When Perry came aboard Niagara, he raised a battle flag with the text "Don't Give Up the Ship" when he ordered the ship to break through the British line which led to the victory for the United States.  Growing up in this city, every year in middle school a field trip was always planned to visit this ship to learn the history and impact it had on the War of 1812.
