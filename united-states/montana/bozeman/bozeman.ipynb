## Bozeman, Montana

## 45° 40′ 40″ N, 111° 2′ 50″ W

## Population: 	114,434

## Landmark: Museum of the Rockies

A museum with a focus on paleontology and history of the Rocky Mountains. Becoming an affliate of the Smithonsian Institution in 2005, the museum has the largest collection of dinosaur remains in the United States, specifically the largest Tyrannosaurus skull. The musuem also has exhibits on Native American cultures on the area, frontier and gold seekers, and even a planetarium.
